```python
df =pd.read_csv('/content/drive/MyDrive/data/titanic (1).csv')
df
```

```python
#En este caso vamos analizar cuantos psajeros hombre y mujeres habian en el Titanic
```

```python
#agrupar columna "sex"
pasajeros = df.groupby('Sex').agg({
    'PassengerId':'count'
})
pasajeros
```

```python
#ahora lo grafico
pasajeros.plot.bar(
    y ='PassengerId',
    width =0.3,
    title='Números de pasajeros por Sexo'
)
```

```python
import seaborn as sns
#ahora sacamos el  indice de  "sex" y la convierto en columna
pasajeros2 =pasajeros.reset_index()
pasajeros2
```

```python
sns.barplot(
    pasajeros2,
    x='Sex',
    y ='PassengerId'
)
```

```python
pasajeros3 =df.groupby('Sex').agg({
    'PassengerId': 'count'

})
```

```python
#Grafico
pasajeros3.plot.pie(
    y = 'PassengerId',
    startangle = 90,
    autopct='%1.1F%%'


)
```

```python
#Hacer gráficos de pie con una librería Plotly
import plotly.express as px
```

```python
# Hacer gráfico
pasajeros3= pasajeros.reset_index()
pasajeros3
```

```python
fig= px.pie(
    pasajeros3,
    names ='Sex',
    values='PassengerId',
    title='Porción de pasajeros del Titanic por sexo'


)
fig.show()
```

```python
#Cargo el DF
df =pd.read_csv('/content/drive/MyDrive/data/titanic (1).csv')
df
```

```python
# Elimino valores nulos en Age
df =df.dropna(subset=['Age'], axis='index')
#diagrama de caja
df.boxplot(
    column='Age'
)
```

```python
#diferenciar por grupo  femenino y masculino
df.boxplot(
    column='Age',
    by  = 'Sex'
)
```

```python
#Vamos a hacer el mismo gráfico pero en seaborn
import seaborn as sns
sns.boxplot(
    df,
    x='Sex',
    y='Age'
)
```

```python
#Gráficos de violines
sns.violinplot(
    df,
    x='Sex',
    y='Age',
    inner='quartites'
)
```
